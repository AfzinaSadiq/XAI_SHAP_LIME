```python
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
from lime import lime_tabular
```

```python
data = load_breast_cancer()
X,y = data['data'],data['target']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
forest_model = RandomForestClassifier()
forest_model.fit(X_train,y_train)
```

```python
forest_model.score(X_test,y_test)
```

```python
explainer = lime_tabular.LimeTabularExplainer(
    training_data=X_train,
    feature_names=data['feature_names'],
    class_names=data['target_names'],
    mode = 'classification'
)
```

```python
for i in range(20):
    print('Correct: ','benign' if y_test[i] else 'malignant')
    print('Classification: ',forest_model.predict([X_test[i]]))
    print(dict(zip(data['feature_names'],X_test[i])))

    explanation = explainer.explain_instance(data_row=X_test[i], predict_fn = forest_model.predict_proba, num_features=30)

    fig = explanation.as_pyplot_figure()
    plt.tight_layout()
    plt.show()
```
